```python
#python3

def python_redis_pub_sub():
    """
    implementation of a pub sub in python using redis 
    """
    import redis
    import time
    import json
    import random
    import threading
    import sys

    def publisher(n):
        time.sleep(1)
        for i in range(n):
            message = { 'foo': i }
            r.publish('channel', json.dumps(message))
            print("Published: {}".format(message))
            time.sleep(1)

    def run_publisher():
        t = threading.Thread(target=publisher, args=(3,))
        t.start()

    def run_subscriber():
        pubsub = r.pubsub()
        pubsub.subscribe(['channel'])
        count = 0
        for item in pubsub.listen():
            print("Received: {}".format(item))
            count += 1
            if count == 4:
                pubsub.unsubscribe()
                print("Unsubscribed")
                break

    r = redis.StrictRedis(host='localhost', port=6379, db=0)
    run_publisher()
    run_subscriber()

if __name__ == "__main__":
    python_redis_pub_sub()
```
